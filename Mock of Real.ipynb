I think it depends on how we receive the data. If its a continuous update, then a line graph might be better, with a normalized estimate to gauge upon. But if its a binary comparison, a pie chart might be prettier. Let's try pie chart first. 

```python
# Data to plot
labels = 'Registered', 'Unregistered'
sizes = [.488, .512]
colors = ['lightGreen', 'Red']
explode = (0.1, 0)  # explode 1st slice

# Plot
plt.pie(sizes, explode=explode, labels=labels, colors=colors,
autopct='%1.1f%%', shadow=True, startangle=140)

plt.title("Before Z", bbox={'facecolor':'0.8', 'pad':2})

plt.axis('equal')
plt.show()

# Data to plot
newsizes = [.8, .2]
colors = ['lightGreen', 'Red']
explode = (0.1, 0)  # explode 1st slice

# Plot
plt.pie(newsizes, explode=explode, labels=labels, colors=colors,
autopct='%1.1f%%', shadow=True, startangle=140)

plt.title("After Z", bbox={'facecolor':'0.8', 'pad':2})

plt.axis('equal')
plt.show()

# Data to plot
labels = 'Registered', 'Unregistered'
sizes = [.999, .001]
colors = ['lightGreen', 'Red']
explode = (0, 0)  # explode 1st slice

# Plot
plt.pie(sizes, explode=explode, labels=labels, colors=colors,
autopct='%1.1f%%', shadow=True, startangle=140)

plt.title("After AFTER Z", bbox={'facecolor':'0.8', 'pad':2})

plt.axis('equal')
plt.show()
```
