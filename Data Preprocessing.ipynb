# 1. Peek at Your Data

```python
# View first 20 rows
url = "https://raw.githubusercontent.com/jbrownlee/Datasets/master/pima-indians-diabetes.data.csv"
names = ['preg', 'plas', 'pres', 'skin', 'test', 'mass', 'pedi', 'age', 'class']
data = pandas.read_csv(url, names=names)
peek = data.head(20)
print(peek)
```

```python
# 2. Dimensions of Your Data

# View dimension of table
shape = data.shape
print(shape)

#so this is 768 patients, 9 different classifiers
```

```python
# 3. Data Types

#See if quantitative, qualitative

types = data.dtypes
print(types)

#so our work is already done, no categorical data, although we already knew that from step 1
```

```python
# 4. Data Summary

pandas.set_option('display.width', 100)
pandas.set_option('precision', 3)
description = data.describe()
print(description)
```

```python
# 5. Correlations
import seaborn as sns
pandas.set_option('precision', 3)
correlations = data.corr(method='pearson')
print(correlations)


#now make a heatmap. I like those more

ax = sns.heatmap(
    correlations, 
    vmin=0, vmax=1, center=0,
    cmap=sns.diverging_palette(20, 220, n=200),
    square=True
)
ax.set_xticklabels(
    ax.get_xticklabels(),
    rotation=45,
    horizontalalignment='right'
);
```

```python
#6 Skewness
skew = data.skew()
print(skew)
```
